```python
df=pd.read_csv("Lung Cancer(in).csv")
```

```python
sns.boxplot(x=df['bmi']) 
plt.title("Boxplot of BMI (Before Outlier Removal)") 
plt.show() 

sns.histplot(df['bmi'], bins=30, kde=True) 
plt.title("BMI Distribution (Before Cleaning)") 
plt.show()
```

```python
# Remove outliers from 'bmi' using IQR 
Q1 = df['bmi'].quantile(0.25) 
Q3 = df['bmi'].quantile(0.75) 
IQR = Q3 - Q1 
# Filter out outliers 
df_clean = df[(df['bmi'] >= Q1 - 1.5 * IQR) & (df['bmi'] <= Q3 + 1.5 * IQR)]
```

```python
sns.boxplot(x=df_clean['bmi']) 
plt.title("Boxplot of BMI (After Outlier Removal)") 
plt.show()
```
